```python
#List2 = [2,70,'work', para, 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 3,10,302.5]
#Multiply each and every element by 2 and display the answer
List2 = [2, 70, 'work', 'para', 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 3, 10, 302.5]

result = []

for item in List2:
    if isinstance(item, (int, float, str, list, tuple)):
        result.append(item * 2)
    elif isinstance(item, set):
        result.append({x * 2 for x in item})
    elif isinstance(item, dict):
        result.append({k * 2: v * 2 if isinstance(v, (int, float, str)) else v
                       for k, v in item.items()})
    else:
        result.append(item)
print(result)
```

```python
data = {
    "book": {
        "Hindi": {"1-4": 60, "5-8": 100, "9-10": 150},
        "Marathi": {"1-4": 60, "5-8": 100, "9-10": 150},
        "English": {"1-4": 80, "5-8": 100, "9-10": 150},
        "Science": {"1-4": 90, "5-8": 120, "9-10": 200},
        "Maths": {"1-4": 100, "5-8": 140, "9-10": 250}
    },
    "notebook": {
        "square": {"100": 40, "200": 70},
        "4 lines": {"100": 30, "200": 50},
        "2 lines": {"100": 30, "200": 50},
        "single lines": {"100": 60, "200": 100},
        "a4": {"100": 100, "200": 180}
    }
}
cart = {}      # stores what user buys
total_bill = 0
while True:
    choice = input("\nEnter book / notebook (or 'stop'): ").lower()

    if choice == "stop":
        break

    if choice == "book":
        subject = input("Enter subject(Hindi/Marathi/English/Science/Maths) : ").title()
        standard = int(input("Enter standard (1–10): "))

        if 1 <= standard <= 4:
            grp = "1-4"
        elif 5 <= standard <= 8:
            grp = "5-8"
        elif 9 <= standard <= 10:
            grp = "9-10"
        else:
            print("Invalid standard")
            continue

        price = data["book"][subject][grp]

        item_name = f"{subject} Book (Std {standard})"
        cart[item_name] = price
        total_bill += price

        print(f"Added: {item_name} – ₹{price}")


    elif choice == "notebook":
        nb_type = input("Enter notebook type (square/4 lines/2 lines/single line/a4 ): ")
        pages = input("Enter pages (100/200): ")

        price = data["notebook"][nb_type][pages]

        item_name = f"{nb_type} Notebook ({pages} pages)"
        cart[item_name] = price
        total_bill += price

        print(f"Added: {item_name} – ₹{price}")

    else:
        print("Invalid choice")


print("\n========== FINAL BILL ==========")

for item, price in cart.items():
    print(f"{item:<25} ₹{price}")

print("--------------------------------")
print(f"Total Amount Payable: ₹{total_bill}")
print("================================")
```

```python
#Create a function to check if string is palindrome or not ?
#For example, if input is NITIN then reverse of the string is same then it is palindrome.
#If input is ANIL then reverse is LINA which is not same then it is not palindrome.

def Palindrome(string):
  if type(string) == 'str':
    if string == string[::-1]:
      return ("palindrome")
    else:
      return ("not a palindrome")
  else:
    return("enter proper string")
```

```python
#Create a function to calculate age till now.
from datetime import datetime

def calculate_age():

    dob = input("Enter DOB (dd-mm-yyyy): ")

    birth_date = datetime.strptime(dob, "%d-%m-%Y")
    today = datetime.today()

    age = today.year - birth_date.year

    if (today.month, today.day) < (birth_date.month, birth_date.day):
        age = age - 1

    print("Your Age is:", age)

calculate_age()
```

```python
#Create a function to check age eligibility for given customer based on DOB. Function will take two input DOB and ELIGIBILITY age.
from datetime import datetime

def check_eligibility(dob, eligibility_age):

    birth_date = datetime.strptime(dob, "%d-%m-%Y")
    today = datetime.today()

    age = today.year - birth_date.year

    if (today.month, today.day) < (birth_date.month, birth_date.day):
        age = age - 1

    if age >= eligibility_age:
        print("Eligible")
    else:
        print("Not Eligible")

dob = input("Enter DOB (dd-mm-yyyy): ")
eligibility_age = int(input("Enter Eligibility Age: "))

check_eligibility(dob, eligibility_age)
```
